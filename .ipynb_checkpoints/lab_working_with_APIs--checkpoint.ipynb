Following the class example, create a function that returns the price, names of origin and arrival airports and the name of the company. Do it for all the flights between two dates that cost the same.

```python
def find_flights_eur(departure_date, return_date):
    import requests

    url = "https://travelpayouts-travelpayouts-flight-data-v1.p.rapidapi.com/v1/prices/calendar"

    querystring = {"calendar_type":"departure_date","origin":"BER","destination":"BCN",
                   "depart_date":str(departure_date),"return_date":str(return_date),"currency":"EUR"}

    headers = {
            "X-Access-Token": "dc4edf46c738c691d13bc643c4ca82cb",
            "X-RapidAPI-Host": "travelpayouts-travelpayouts-flight-data-v1.p.rapidapi.com",
            "X-RapidAPI-Key": "71fd85f769mshdae97e61eb4dcb0p17c234jsn29ada1c2d0e7"
    }

    response = requests.request("GET", url, headers=headers, params=querystring)
    dates_of_flights = list(response.json()['data'].keys())
    prices = list(response.json()['data'].keys())
    flights = pd.DataFrame(columns=['origin', 'destination', 'price', 'airline', 'flight_number',
       'departure_at', 'return_at', 'transfers', 'expires_at'])
    for i in range(len(dates_of_flights)):
        flights = flights.append(pd.json_normalize(response.json()['data'][dates_of_flights[i]]))
    flights = flights.drop(['flight_number', 'transfers', 'expires_at'], axis = 1)
    
    return flights
```

```python
flight=find_flights_eur('2022-04', '2022-05')
flight
```

```python
flight_grouped = flight.groupby('price')
flight_price = flight_grouped['origin', 'destination','airline', 'departure_at', 'return_at'].apply(list).reset_index()
print(flight_price)
```
